```python
import pandas as pd

# Sample DataFrame
data = {
    'year': [1972, 1973, 1974, 1975, 1976],
    'id': [1, 2, 3, 4, 5],
    1972: [-100, 2, 2, 2, 2],
    1973: [2, 2, 2, 2, 2],
    1974: [2, 2, 2, 2, -100],
    1975: [-100, -100, 2, -100, -100],
    1976: [2, 2, 2, 2, 2]
}
df = pd.DataFrame(data)

# Replace values with their meanings
df.replace({
    -100: 'inapplicable',
    1: 'yes',
    2: 'no'
}, inplace=True)

# Melt the DataFrame to have years and values in separate rows
melted_df = df.melt(id_vars=['id', 'year'], var_name='year_col', value_name='value')

# Get value counts for each year and value
value_counts_df = melted_df.groupby(['year_col', 'value']).size().unstack(fill_value=0)

# Pivot the DataFrame so that years become columns
value_counts_df = value_counts_df.transpose()

# Display the counts DataFrame
print(value_counts_df)
```
