```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.concat( 
    map(pd.read_csv, ['fixedGSSeverdivorced.csv', 'fixedGSShappinessofmarriage.csv', "fixedGSShighestschoolspouse.csv",'fixedGSSincome.csv', 'fixedGSSmaritalstatus.csv', 'fixedGSSnumberofchildren.csv','fixedGSSspouselaborstatus.csv']), ignore_index=True) 
print(df)
```

```python
# drop all negative values from final df 
value_to_drop = -100.0
value_to_drop1 = -99.0
value_to_drop2 = -98.0
value_to_drop3 = -97.0

mask = df['code_divorced'] == value_to_drop
mask1 = df['code_divorced'] == value_to_drop1
mask2 = df['code_divorced'] == value_to_drop2
mask3 = df['code_divorced'] == value_to_drop3

df_filtered = df.drop(df[mask, mask1, mask2, mask3].index)

print(df_filtered)
```

```python
# Values to drop
values_to_drop = [-100.0, -99.0, -98.0, -97.0]

# Create a single mask to filter rows
mask = df['code_divorced'].isin(values_to_drop)

# Invert the mask to keep rows that are not in the values_to_drop list
df_filtered = df[~mask]

print(df_filtered)
```

```python
# Values to drop
values_to_drop = [-100.0, -99.0, -98.0, -97.0]

# Create a single mask to filter rows
mask = df_filtered['label_divorced'].isin(values_to_drop)

# Invert the mask to keep rows that are not in the values_to_drop list
df_filtered = df_filtered[~mask]

df_filtered.head()
```
